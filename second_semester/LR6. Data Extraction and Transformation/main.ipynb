```python
# статистика по пустым столбцам

blanks_statistics = pd.DataFrame({
    'Всего': df.isna().sum(),
    'Процент пропущенных пробелов': (df.isna().sum() / df.shape[0] * 100)
})   

blanks_statistics.sort_values(by='Процент пропущенных пробелов', ascending=False)
```
