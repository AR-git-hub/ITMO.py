```python
# Импорт всех модулей
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
```

```python
# Настройки
# Настройки
pd.set_option('display.max_columns', None)
sns.set_theme(style='whitegrid', palette='Set2')
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12

plots_dir = 'plots'
os.makedirs(plots_dir, exist_ok=True)
```

## Первичный анализ

Загрузка данных и базовая проверка

```python
# Импорт данных

df = pd.read_csv('data/titanic_dataset.csv')

df.head(10)
# df.tail(10)
```

```python
# Проверка типов данных каждого столбца
df.dtypes
```

```python
# Проверка общей информации и типов данных колонок
df.info()
```

```python
# Статистика пропусков
blanks_statistics = pd.DataFrame({
    'Всего': df.isna().sum(),
    'Доля': (df.isna().sum() / df.shape[0] * 100).round(2)
})
blanks_statistics = blanks_statistics.sort_values(by='Доля', ascending=False)

# График пропусков
plt.figure(figsize=(10, 4))
blanks_statistics['Доля'].plot(kind='bar', color='#66c2a5')
plt.title('Доля пропусков по столбцам')
plt.ylabel('Проценты')
plt.tight_layout()
plt.savefig(f'{plots_dir}/01_missing_values.png', dpi=150)
plt.show()

blanks_statistics
```

```python
(df.isna().sum() / df.shape[0] * 100).values
```

```python
# статистическая информация про каждый числовой столбец
df.describe().T
```

```python
# Гистограммы числовых признаков
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols].hist(figsize=(14, 10), bins=20)
plt.tight_layout()
plt.savefig(f'{plots_dir}/02_numeric_histograms.png', dpi=150)
plt.show()
```

```python
# Пропуски по столбцам
missing_before = df.isna().sum()
missing_before[missing_before > 0].sort_values(ascending=False)
```

```python
# Заполнение Age и новый признак
age_mean = df['Age'].mean()
age_median = df['Age'].median()
df['Age'] = df['Age'].fillna(age_median)

df['Age_group'] = pd.cut(
    df['Age'],
    bins=[0, 12, 18, 35, 60, np.inf],
    labels=['Ребенок', 'Подросток', 'Взрослый', 'Зрелый', 'Пожилой'],
    include_lowest=True
)

age_mean, age_median
```

```python
# Заполнение Embarked и признак палубы
embarked_mode = df['Embarked'].mode().iloc[0]
df['Embarked'] = df['Embarked'].fillna(embarked_mode)
df['CabinDeck'] = df['Cabin'].fillna('Unknown').str[0]

embarked_mode
```

```python
# Трансформация типов и новые признаки
df['Pclass'] = df['Pclass'].replace({1: 'F', 2: 'S', 3: 'T'})
df['Pclass'] = df['Pclass'].astype('category')

# Титул из имени
df['Title'] = df['Name'].str.extract(r',\s*([^\.]+)\.', expand=False).str.strip()

# Пол в числовой вид
df['Sex'] = df['Sex'].replace({'female': 0, 'male': 1}).astype(int)

# Размер семьи и одиночка
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)

df[['Pclass', 'Title', 'Sex', 'FamilySize', 'IsAlone']].head()
```

```python
# Boxplot для Fare
plt.figure(figsize=(10, 4))
sns.boxplot(x=df['Fare'])
plt.title('Boxplot Fare')
plt.tight_layout()
plt.savefig(f'{plots_dir}/03_fare_boxplot.png', dpi=150)
plt.show()

# Выбросы IQR
q1 = df['Fare'].quantile(0.25)
q3 = df['Fare'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
fare_outliers = df[(df['Fare'] < lower_bound) | (df['Fare'] > upper_bound)]

fare_outliers[['PassengerId', 'Fare']].head(), fare_outliers.shape[0]
```

```python
# Распределение Age
plt.figure(figsize=(10, 4))
sns.histplot(df['Age'], kde=True, bins=25)
plt.title('Распределение Age')
plt.tight_layout()
plt.savefig(f'{plots_dir}/04_age_distribution.png', dpi=150)
plt.show()

# Winsorize для Fare
df['Fare_winsor'] = sp.stats.mstats.winsorize(df['Fare'], limits=(0.05, 0.05)).astype(float)

# Ограничение Age по 95 перцентилю
age_p95 = df['Age'].quantile(0.95)
df['Age_capped'] = df['Age'].clip(upper=age_p95)

age_p95
```

## Обработка и трансформация

```python
# Агрегации
survival_by_pclass = df.groupby('Pclass', observed=False)['Survived'].mean().sort_index()

group_pclass_sex = (
    df.groupby(['Pclass', 'Sex'], observed=False)['Survived']
    .mean()
    .rename('survived_rate')
)

median_age_by_embarked = df.groupby('Embarked', observed=False)['Age_capped'].median().sort_index()

survival_pivot = pd.pivot_table(
    df,
    values='Survived',
    index='Age_group',
    columns='IsAlone',
    aggfunc='mean'
)

survival_by_pclass, group_pclass_sex, median_age_by_embarked, survival_pivot
```

```python
# Метрики качества
missing_after = df.isna().sum()
filled_count = (missing_before - missing_after).clip(lower=0).sum()
missing_filled_percent = round(filled_count / missing_before.sum() * 100, 2)

categorical_cols = ['Pclass', 'Embarked', 'CabinDeck', 'Title', 'Age_group']
unique_counts = df[categorical_cols].nunique(dropna=False)

pclass_dist = df['Pclass'].value_counts(normalize=True).round(3)
sex_dist = df['Sex'].value_counts(normalize=True).round(3)
isalone_dist = df['IsAlone'].value_counts(normalize=True).round(3)

corr_matrix = df[['Sex', 'FamilySize', 'IsAlone', 'Fare_winsor', 'Age_capped', 'Survived']].corr(numeric_only=True)

missing_filled_percent, unique_counts, pclass_dist, sex_dist, isalone_dist, corr_matrix
```

```python
# Сохранение очищенных данных
output_path = 'data/titanic_cleaned.csv'
df.to_csv(output_path, index=False)

saved_plots = [
    'plots/01_missing_values.png',
    'plots/02_numeric_histograms.png',
    'plots/03_fare_boxplot.png',
    'plots/04_age_distribution.png'
]

df.shape, output_path, saved_plots
```
